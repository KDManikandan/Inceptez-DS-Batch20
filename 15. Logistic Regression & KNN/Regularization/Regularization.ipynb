```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
df = pd.read_csv("test.csv",names = ['X1','Y'])
```

```python
df
```

```python
model = LinearRegression()
model.fit(df[['X1']],df['Y'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model = LinearRegression()
model.fit(df[['X1']],df['Y'])

print(model.coef_)
print(model.intercept_)

pred_df = model.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'Blue')
```

```python
model_ = Ridge(alpha = 0.05)
model_.fit(df[['X1']],df['Y'])

print(model_.coef_)
print(model_.intercept_)

pred_df = model.predict(df[['X1']])
pred_df_ = model_.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'Blue')
plt.plot(df.X1,pred_df_,'green')
```

```python
model_ = Lasso(alpha = 0.05)
model_.fit(df[['X1']],df['Y'])

#Lasso - L1
#Ridge - L2
print(model_.coef_)
print(model_.intercept_)

pred_df = model.predict(df[['X1']])
pred_df_ = model_.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'Blue')
plt.plot(df.X1,pred_df_,'green')
```

```python
for i in np.arange(0,0.05,0.001):
  print("i value: ", i)
  model_ = Lasso(alpha = i)
  model_.fit(df[['X1']],df['Y'])

  print(model_.coef_)
  print(model_.intercept_)

  pred_df = model.predict(df[['X1']])
  pred_df_ = model_.predict(df[['X1']])
  plt.scatter(df.X1,df.Y,color = 'red')
  plt.plot(df.X1,pred_df,'Blue')
  plt.plot(df.X1,pred_df_,'green')
```

```python
model_ = ElasticNet(l1_ratio = 0.2,alpha =0.003)
model_.fit(df[['X1']],df['Y'])

print(model_.coef_)
print(model_.intercept_)

pred_df = model.predict(df[['X1']])
pred_df_ = model_.predict(df[['X1']])
plt.scatter(df.X1,df.Y,color = 'red')
plt.plot(df.X1,pred_df,'Blue')
plt.plot(df.X1,pred_df_,'green')
```
